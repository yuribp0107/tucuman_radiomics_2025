##**Extraccion de Caracteristicas de Radiomics**

```python
# IMPORTANTE: Ejecutar esta celda para importar los datos desde Kaggle

import kagglehub  # Biblioteca para descargar datasets de Kaggle de forma sencilla

# Descarga el conjunto de datos 'chest-xray-pneumonia' del usuario 'paultimothymooney'
paultimothymooney_chest_xray_pneumonia_path = kagglehub.dataset_download('paultimothymooney/chest-xray-pneumonia')

print('Importación del conjunto de datos completada.')

# Define las rutas a los subconjuntos de validación y prueba del dataset
val_dir = os.path.join("/kaggle/input/chest-xray-pneumonia/chest_xray", "val")
test_dir = os.path.join("/kaggle/input/chest-xray-pneumonia/chest_xray", "test")
```

```python
# --- Función de segmentación robusta ---
def get_lung_mask_sitk(img: sitk.Image) -> sitk.Image:
    img_norm = sitk.RescaleIntensity(img, 0.0, 1.0)
    otsu = sitk.OtsuThresholdImageFilter()
    otsu.SetInsideValue(1); otsu.SetOutsideValue(0)
    otsu.SetNumberOfHistogramBins(128)
    mask = otsu.Execute(img_norm)
    mask = sitk.BinaryMorphologicalClosing(mask, kernelRadius=(50,50), foregroundValue=1)
    mask = sitk.BinaryFillhole(mask, fullyConnected=True)

    cc    = sitk.ConnectedComponent(mask)
    stats = sitk.LabelShapeStatisticsImageFilter(); stats.Execute(cc)
    labels = stats.GetLabels()

    if len(labels) >= 2:
        l1, l2 = sorted(labels, key=lambda l: stats.GetPhysicalSize(l), reverse=True)[:2]
        lung_mask = (sitk.BinaryThreshold(cc, l1, l1, 1, 0)
                   + sitk.BinaryThreshold(cc, l2, l2, 1, 0))
    elif len(labels) == 1:
        lung_mask = sitk.BinaryThreshold(cc, labels[0], labels[0], 1, 0)
    else:
        # Fallback robusto: background (0) en el borde, ROI (1) en el centro
        arr = np.ones_like(sitk.GetArrayFromImage(img), dtype=np.uint8)
        arr[0, :] = arr[-1, :] = arr[:, 0] = arr[:, -1] = 0
        lung_mask = sitk.GetImageFromArray(arr)

    lung_mask.CopyInformation(img)
    return lung_mask

# --- Configuración de PyRadiomics ---
settings = {'binWidth':25, 'resampledPixelSpacing':None, 'interpolator':'sitkBSpline'}
extractor = featureextractor.RadiomicsFeatureExtractor(**settings)
extractor.enableAllFeatures()

# --- Función de extracción con manejo de errores ---
def extraer_features(dir_base: str, clases: list) -> pd.DataFrame:
    rows = []
    for clase in clases:
        path = os.path.join(dir_base, clase)
        if not os.path.isdir(path): continue
        for fname in os.listdir(path):
            if not fname.lower().endswith(('.png','.jpg','.jpeg')): continue

            img  = sitk.ReadImage(os.path.join(path, fname), sitk.sitkFloat32)
            mask = get_lung_mask_sitk(img)

            # Asegurar dos etiquetas en la máscara
            arr = sitk.GetArrayFromImage(mask)
            if np.unique(arr).size == 1:
                full = np.ones_like(arr, dtype=np.uint8)
                full[0, :] = full[-1, :] = full[:, 0] = full[:, -1] = 0
                mask = sitk.GetImageFromArray(full)
                mask.CopyInformation(img)

            # Extraer características o saltar si falla
            try:
                feats = extractor.execute(img, mask)
            except ValueError as e:
                print(f"[SKIP] {clase}/{fname}: {e}")
                continue

            feats['Clase']    = clase
            feats['Filename'] = fname
            rows.append(feats)

    return pd.DataFrame(rows)

# --- Llamadas a cada split y concatenación ---
if __name__ == "__main__":
    clases     = ['NORMAL', 'PNEUMONIA']
    # Ajusta estas rutas a tu estructura
    val_dir    = "/kaggle/input/chest-xray-pneumonia/chest_xray/val"
    test_dir   = "/kaggle/input/chest-xray-pneumonia/chest_xray/test"

    df_val   = extraer_features(val_dir,   clases)
    df_test  = extraer_features(test_dir,  clases)

    # Añadir columna para saber el origen
    df_val  ['Subset'] = 'val'
    df_test ['Subset'] = 'test'

    # Concatenar sin perder el flujo
    df_all = pd.concat([df_val, df_test],
                       axis=0,
                       ignore_index=True)

    # Guardar resultado
    df_all.to_csv("radomics_all_splits.csv", index=False)
df_all
```

2. y_encoded = le.fit_transform(y): Convierte la columna y (que contiene las clases) en un vector numérico. Por ejemplo, puede asignar NORMAL → 0 y PNEUMONIA → 1.
3. print(...): Muestra en pantalla cómo se realizó el mapeo, es decir, qué número corresponde a cada clase.

####**Dividir los datos en conjunto de entrenamiento y prueba**

Esta línea de código divide los datos en un conjunto de entrenamiento (X_train y y_train) y un conjunto de prueba (X_test y y_test) de acuerdo con las proporciones especificadas. El conjunto de entrenamiento se utiliza para entrenar el modelo, y el conjunto de prueba se utiliza para evaluar qué tan bien el modelo generaliza a datos no vistos.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,                   # Matriz de características (variables independientes)
    y_encoded,           # Vector de etiquetas codificadas (variable objetivo)
    test_size=0.2,       # Fracción de datos reservada para el conjunto de prueba (20%)
    random_state=42,     # Semilla para garantizar que la división sea reproducible
    stratify=y_encoded   # Mantiene la proporción de clases en ambos subconjuntos
)
```

3. train_test_split(...): Esta función de scikit-learn divide los datos de forma aleatoria y controlada en entrenamiento y prueba. Sus parámetros son:
- X: Matriz de características (variables independientes).
- y_encoded: Vector de etiquetas codificadas (variable objetivo).
- test_size=0.2: Asigna el 20% de los datos al conjunto de prueba, y el 80% restante al conjunto de entrenamiento.
- random_state=42: Fija una semilla aleatoria para garantizar que la división sea reproducible en futuras ejecuciones.
- stratify=y_encoded: Asegura que la proporción de clases (normal vs neumonía) se mantenga equilibrada tanto en el conjunto de entrenamiento como en el de prueba. Esto es importante para evitar sesgos durante la evaluación del modelo.

####**Definir el Precocesamiento**

```python
# Este pipeline encadena varios pasos de preprocesamiento y modelado para facilitar la reproducibilidad y limpieza del código
pipeline = ImbPipeline([

    # 1. Normalización de las características a un rango entre 0 y 1
    ('scaler', MinMaxScaler()),

    # 2. Aplicación de SMOTE (Synthetic Minority Over-sampling Technique), para balancear las clases
    ('smote', SMOTE(random_state=42)),

    # 3. Selección de características más relevantes(por defecto, ANOVA F-test)
    ('select', SelectKBest()),

    # 4. Clasificador: Árbol de decisión
    ('clf', DecisionTreeClassifier(random_state=42))
])
```

1. MinMaxScaler(): Normaliza los valores de cada característica para que estén en el rango [0, 1]. Esto asegura que todas las variables contribuyan de forma equitativa al análisis, especialmente importante para algoritmos sensibles a la escala como SMOTE y algunos clasificadores.

2. SMOTE() (Synthetic Minority Over-sampling Technique): Genera ejemplos sintéticos de la clase minoritaria (por ejemplo, neumonía si está subrepresentada) con el fin de balancear el conjunto de entrenamiento. Esto mejora el desempeño del modelo en escenarios con clases desbalanceadas.

3. SelectKBest(): Selecciona automáticamente las características más relevantes del conjunto de datos en función de un criterio estadístico (por defecto, ANOVA F-test). Esto reduce la dimensionalidad y mejora la eficiencia del modelo.

4. DecisionTreeClassifier(): Entrena un árbol de decisión utilizando las características seleccionadas. Este modelo genera reglas de clasificación interpretables, útiles para entender qué variables radiomics ayudan a distinguir entre radiografías normales y con neumonía.

####**Definir la cuadrícula de parámetros para seleccionar el "mejor" modelo**

```python
# Diccionario de hiperparámetros que se evaluarán en la búsqueda por cuadrícula (GridSearchCV)
param_grid = {

    # Número de vecinos que SMOTE utilizará para generar muestras sintéticas.
    'smote__k_neighbors': [3, 5],

    # Número de características que se seleccionarán en el paso de SelectKBest.
    'select__k': [5, 10, 'all'],

    # Criterio de división para el árbol de decisión:
    # 'gini' (impureza de Gini) o 'entropy' (ganancia de información).
    'clf__criterion': ['gini', 'entropy'],

    # Profundidad máxima del árbol.
    'clf__max_depth': [3, 5],

    # Número mínimo de muestras necesarias para dividir un nodo interno.
    'clf__min_samples_split': [2, 5]
}
```

1. El diccionario param_grid define un conjunto de hiperparámetros que serán evaluados durante la búsqueda de la mejor combinación mediante GridSearchCV.
2. El parámetro smote__k_neighbors ajusta el número de vecinos que SMOTE utiliza para generar muestras sintéticas de la clase minoritaria, lo cual influye en la diversidad de los datos balanceados.
3. El parámetro select__k controla cuántas características serán seleccionadas por SelectKBest, lo que impacta directamente en la dimensionalidad del conjunto de datos y en la capacidad del modelo para generalizar.
4. Por su parte, clf__criterion define el criterio que el árbol de decisión usará para dividir los nodos, ya sea mediante impureza de Gini o ganancia de información (entropía).
5. El parámetro clf__max_depth establece la profundidad máxima del árbol, lo que permite controlar el sobreajuste al limitar la complejidad del modelo.
6. Finalmente, clf__min_samples_split indica el número mínimo de muestras necesarias para dividir un nodo interno, ayudando a regular la estructura del árbol y su capacidad de adaptación a los datos de entrenamiento.

####**Configuracion de la validación cruzada**

Esta línea de código se utiliza para configurar la validación cruzada estratificada en el contexto del aprendizaje automático. La validación cruzada es una técnica que se utiliza para evaluar el rendimiento de un modelo de manera más confiable al dividir el conjunto de datos en múltiples particiones y realizar múltiples iteraciones de entrenamiento y evaluación.

```python
#La validación cruzada estratificad
CV =RepeatedStratifiedKFold(n_splits=10, n_repeats=1, random_state=42)
```

1.   n_splits=10: Este parámetro especifica el número de divisiones (o "pliegues") que se crearán en la validación cruzada. En este caso, se crean 10 divisiones, lo que significa que el conjunto de datos se dividirá en 10 partes iguales.
2.   n_repeats=1: Este parámetro especifica el número de repeticiones de la validación cruzada. Una repetición significa que se realizará el proceso de división y evaluación del modelo múltiples veces. En este caso, se realiza una sola repetición.
3.   random_state=42: Este parámetro establece una semilla para la generación de números aleatorios. La semilla asegura que la validación cruzada sea reproducible, es decir, si se utiliza la misma semilla en diferentes ejecuciones, se obtendrá la misma división de datos. Esto es útil para garantizar que los resultados sean coherentes en diferentes ejecuciones.

La validación cruzada estratificada es una variante de la validación cruzada que asegura que cada pliegue (fold) mantenga la misma proporción de clases que el conjunto de datos original. Esto es especialmente importante en problemas de clasificación donde las clases pueden estar desbalanceadas. El uso de RepeatedStratifiedKFold en lugar de KFold o StratifiedKFold implica que se realizará la validación cruzada estratificada varias veces (en este caso, una sola repetición) para obtener estimaciones de rendimiento más confiables del modelo.



####**Configurar GridSearchCV**

```python
# Se crea un objeto GridSearchCV para realizar búsqueda exhaustiva de hiperparámetros en el pipeline definido
grid = GridSearchCV(

    pipeline,          # El pipeline que contiene los pasos: escalado, SMOTE, selección de características y clasificador

    param_grid,        # Diccionario con los hiperparámetros que se evaluarán (definido previamente)

    cv=CV,             # Estrategia de validación cruzada, por ejemplo RepeatedStratifiedKFold para mantener equilibrio entre clases

    scoring='roc_auc', # Métrica de evaluación utilizada: área bajo la curva ROC, ideal para clasificación binaria y datos desbalanceados

    n_jobs=-1,         # Utiliza todos los núcleos disponibles del procesador para paralelizar la búsqueda (acelera el proceso)

)
```

- Este bloque de código crea un objeto GridSearchCV, el cual se encarga de buscar automáticamente la mejor combinación de hiperparámetros para un pipeline de aprendizaje automático.
- Utiliza un conjunto predefinido de configuraciones (param_grid) y evalúa cada una mediante validación cruzada estratificada (cv=CV), asegurando que la proporción entre clases (normal y neumonía) se mantenga en cada partición.
- La métrica de evaluación utilizada es el área bajo la curva ROC (scoring='roc_auc'), especialmente útil en problemas de clasificación binaria con clases desbalanceadas. Además, el proceso se paraleliza usando todos los núcleos del procesador (n_jobs=-1), lo que acelera significativamente la búsqueda.

####**Entrenar el modelo**

```python
# Mensaje informativo para indicar que comienza el proceso de entrenamiento y búsqueda de hiperparámetros
print("Entrenando el modelo con GridSearchCV...")

# Se entrena el pipeline utilizando las distintas combinaciones de parámetros definidas en param_grid
# GridSearchCV evaluará cada combinación mediante validación cruzada y seleccionará la mejor
grid.fit(X_train, y_train)

# Se imprime la mejor combinación de hiperparámetros encontrada durante la validación cruzada
print(f"\nMejores parámetros (CV): {grid.best_params_}")

# Se imprime el mejor valor del área bajo la curva ROC (AUC) obtenido durante la validación cruzada
print(f"Mejor AUC CV: {grid.best_score_:.3f}")
```

Este bloque de código entrena un modelo de clasificación utilizando GridSearchCV, una herramienta que permite buscar automáticamente la mejor combinación de hiperparámetros dentro de un pipeline previamente definido
1. Primero, imprime un mensaje indicando que comienza el entrenamiento. Luego, con grid.fit(X_train, y_train), ejecuta el proceso de validación cruzada, evaluando múltiples configuraciones del modelo sobre los datos de entrenamiento.
2. Al finalizar, se muestran en pantalla los mejores parámetros encontrados (grid.best_params_) y el mejor valor del área bajo la curva ROC (grid.best_score_), lo que indica qué configuración obtuvo el mejor rendimiento promedio en la clasificación entre neumonía y normal durante la validación cruzada.

####**Evaluacion del modelo**

```python
# Imprime un mensaje indicando que comienza la evaluación del modelo final sobre el conjunto de prueba
print("\nEvaluando el mejor modelo en el conjunto de prueba...")

# Se recupera el mejor modelo (pipeline completo) encontrado por GridSearchCV
best_model = grid.best_estimator_

# Se generan las predicciones de clase para el conjunto de prueba
y_pred = best_model.predict(X_test)

# Se obtienen las probabilidades predichas de la clase positiva (neumonía)
# Esto es útil para calcular métricas basadas en probabilidad como el AUC
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

# Se calcula un conjunto de métricas para evaluar el rendimiento del modelo sobre el conjunto de prueba
metrics = {
    'accuracy': accuracy_score(y_test, y_pred),  # Porcentaje de predicciones correctas
    'precision': precision_score(y_test, y_pred, pos_label=pos_label_encoded),  # Precisión para la clase positiva (neumonía)
    'recall': recall_score(y_test, y_pred, pos_label=pos_label_encoded),        # Sensibilidad o TPR para la clase positiva
    'f1': f1_score(y_test, y_pred, pos_label=pos_label_encoded),                # Media armónica entre precisión y recall
    'auc': roc_auc_score(y_test, y_pred_proba)                                  # Área bajo la curva ROC (métrica robusta para desbalance)
}

# Se imprimen las métricas calculadas para evaluar el desempeño final del modelo
print(f"Métricas en el conjunto de prueba:")
for metric, value in metrics.items():
    print(f"  {metric.capitalize()}: {value:.3f}")

```

Este bloque de código evalúa el rendimiento del mejor modelo encontrado por GridSearchCV sobre el conjunto de prueba.
1. Primero, se imprime un mensaje informativo para indicar el inicio de la evaluación.
2. Luego, se extrae el mejor estimador (best_model) encontrado durante la validación cruzada.
3. A continuación, se generan las predicciones de clase (y_pred) y las probabilidades de pertenencia a la clase positiva (y_pred_proba), que en este caso corresponde a neumonía.
4. Posteriormente, se calculan cinco métricas clave para evaluar el rendimiento del modelo: accuracy (exactitud general), precision (proporción de verdaderos positivos entre todos los positivos predichos), recall (sensibilidad o tasa de verdaderos positivos), f1-score (media armónica entre precisión y recall) y AUC (área bajo la curva ROC, útil para evaluar la discriminación del modelo en problemas desbalanceados).
5. Finalmente, se imprimen los resultados de estas métricas, lo que permite interpretar de forma clara el desempeño del modelo al clasificar entre neumonía y normal en datos no vistos.

```python
# Informe de clasificación
print("Informe de Clasificación:")

# Se imprime un resumen de métricas de desempeño del modelo utilizando classification_report de scikit-learn
# Incluye precisión, recall, F1-score y soporte para cada clase (NORMAL y PNEUMONIA)
print(classification_report(y_test, y_pred, target_names=["NORMAL", "PNEUMONIA"]))
```

- precisión (proporción de predicciones correctas entre las realizadas para una clase),
- recall (capacidad del modelo para detectar correctamente los casos reales de una clase),
- F1-score (media armónica entre precisión y recall) y
- el soporte (número real de muestras por clase).

Este informe es especialmente útil para comprender cómo se comporta el modelo con cada categoría por separado, lo cual es fundamental en entornos médicos donde el costo de un error puede variar según la clase (por ejemplo, un falso negativo en neumonía puede ser clínicamente más grave que un falso positivo).

```python
def get_classifiers(random_state=42):
    """
    Devuelve un diccionario de clasificadores para comparar su rendimiento
    tras aplicar LASSO (Selección L1) y MinMaxScaler.

    Args:
        random_state (int): Semilla para reproducibilidad en los clasificadores que la soporten.

    Returns:
        dict: Mapa nombre -> instancia de clasificador.
    """
    return {
        # Clasificador de regresión logística con regularización L1 (LASSO)
        "Logistic_L1": LogisticRegression(
            penalty='l1', solver='liblinear', C=1.0, random_state=random_state
        ),

        # Árbol de decisión sin límite de profundidad
        "DecisionTree": DecisionTreeClassifier(
            max_depth=None, random_state=random_state
        ),

        # Bosque aleatorio con 100 árboles
        "RandomForest": RandomForestClassifier(
            n_estimators=100, random_state=random_state
        ),

        # Máquina de soporte vectorial con kernel RBF y probabilidad habilitada
        "SVM_RBF": SVC(
            kernel='rbf', probability=True, random_state=random_state
        ),

        # Clasificador de K vecinos más cercanos con k=5
        "KNN": KNeighborsClassifier(
            n_neighbors=5
        ),

        # Clasificador probabilístico Naive Bayes Gaussiano
        "NaiveBayes": GaussianNB(),

        # Clasificador Perceptrón
        "Perceptron": Perceptron(max_iter=1000, tol=1e-3, random_state=random_state),

         # Perceptrón multicapa (red neuronal simple) con 1 capa oculta de 100 neuronas
        "MLP": MLPClassifier(
            hidden_layer_sizes=(100,), max_iter=500, random_state=random_state
        ),

        # Clasificador XGBoost con métrica logloss y semilla fija
        "XGBoost": XGBClassifier(
            eval_metric='logloss', random_state=random_state
        )
    }
```

La función get_classifiers() construye un conjunto de clasificadores populares del aprendizaje automático y los devuelve en forma de diccionario.
Cada clasificador está configurado con una semilla aleatoria (random_state) para asegurar reproducibilidad. Entre los modelos incluidos se encuentran:
- regresión logística con penalización L1 (LASSO),
- árbol de decisión,
- random forest,
- máquina de soporte vectorial con kernel RBF,
- K vecinos más cercanos, Naive Bayes gaussiano,
- red neuronal tipo perceptrón multicapa (MLP),
- y el potente algoritmo XGBoost.

Esta función está diseñada para facilitar la comparación del rendimiento de estos clasificadores cuando se aplican técnicas previas como normalización con MinMaxScaler y selección de características con LASSO, dentro de un flujo de trabajo automatizado y reproducible.

####**Experimento "Clasificadores"**

```python
# Supón que get_classifiers() está definida y devuelve tu diccionario de clasificadores
classifiers = get_classifiers(random_state=42)

# Definimos LASSO selector (regularización L1) para filtrar características
lasso_selector = SelectFromModel(
    LogisticRegression(penalty='l1', solver='liblinear', C=1.0, random_state=42),
    prefit=False,
    max_features=10,
    threshold=-np.inf
)

# Estrategia de validación cruzada estratificada
CV = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=42)

plt.figure(figsize=(8, 6))

# Comparamos cada clasificador con GridSearchCV, usando MinMaxScaler + LASSO como preprocessing
for name, clf in classifiers.items():
    # 1) Construimos el pipeline: escala → selección L1 → clasificador
    pipeline = ImbPipeline([
        ("scaler", MinMaxScaler()),
        ("selector", lasso_selector),
        ("clf", clf)
    ])

    # 2) Definimos GridSearchCV para evaluar este único clasificador
    param_grid = {"clf": [clf]}
    grid = GridSearchCV(
        pipeline,
        param_grid,         # Solo varía el clasificador
        cv=CV,              # Validación cruzada estratificada
        scoring="roc_auc",  # Métrica AUC para problemas desbalanceados
        n_jobs=-1,          # Paraleliza en todos los núcleos
        verbose=0
    )

    # 3) Ajustamos el modelo sobre el conjunto de entrenamiento
    grid.fit(X_train, y_train)


    # 4) Obtenemos un "score" continuo para la clase positiva:
    if hasattr(grid, "predict_proba"):
        y_score = grid.predict_proba(X_test)[:, 1]
    else:
        # Perceptron, SVM sin probabilidad, etc.
        y_score = grid.decision_function(X_test)

    # 5) Usamos ese score con roc_curve/roc_auc_score
    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc_score = roc_auc_score(y_test, y_score)

    # 6) Graficamos la curva ROC para este clasificador
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc_score:.3f})")

# Línea diagonal de referencia (clasificador aleatorio)
plt.plot([0, 1], [0, 1], "--", color="gray", alpha=0.5, label="Aleatorio")

# Etiquetas y título
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curvas ROC: Comparación de Clasificadores con LASSO + MinMaxScaler")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

Este fragmento de código compara el rendimiento de distintos algoritmos de clasificación (como regresión logística, árboles de decisión, random forest, SVM, KNN, Naive Bayes, MLP y XGBoost) aplicando un mismo esquema de preprocesamiento: normalización de características con MinMaxScaler y selección de las variables más relevantes mediante LASSO (regularización L1).

Para cada clasificador, se construye un pipeline y se evalúa mediante GridSearchCV con validación cruzada estratificada y métrica ROC AUC, lo que garantiza una evaluación justa y reproducible del desempeño del modelo.

Finalmente, se grafican las curvas ROC de cada clasificador en un mismo gráfico, permitiendo visualizar comparativamente la capacidad discriminativa de cada algoritmo bajo el mismo conjunto de condiciones. Esta estrategia es útil para determinar cuál modelo se adapta mejor a los datos cuando se combinan técnicas específicas de preprocesamiento.

###**Comparando metodos de seleccion de caracteristicas**

####**Métodos de seleccion de caracteristicas usados en el experimento**

```python
def get_feature_selection_methods(nFeatures):
    return {
        # Método embedded supervisado (regularización L1)
        "LASSO": SelectFromModel(
            LogisticRegression(penalty='l1', max_iter=1000, solver='liblinear', C=5, random_state=42),
            prefit=False,
            max_features=nFeatures,
            threshold=-np.inf
        ),

        # Método estadístico tipo filtro (ANOVA F-value)
        "Anova": SelectKBest(k=nFeatures),

        # Método basado en modelo de ensamblado (ExtraTrees)
        "ET": SelectFromModel(
            ExtraTreesClassifier(random_state=42),
            prefit=False,
            max_features=nFeatures,
            threshold=-np.inf
        ),

        # Reducción de dimensionalidad no supervisada
        "PCA": PCA(n_components=nFeatures)
    }
```

Este fragmento de código define una función llamada *get_feature_selection_methods* que devuelve un diccionario con tres métodos distintos de selección de características, permitiendo elegir las variables más relevantes de un conjunto de datos para tareas de aprendizaje automático.

Cada método aplica un enfoque diferente:
- *LASSO* utiliza regresión logística con regularización L1 para seleccionar características con coeficientes distintos de cero,
- *Anova* emplea pruebas estadísticas univariadas (como F-test) para elegir las mejores variables según su relación con la clase objetivo, y
- *ET (Extra Trees)* aplica un modelo de ensamble basado en árboles para estimar la importancia de cada variable.
- *PCA (Análisis de Componentes Principales)* es una técnica no supervisada de reducción de dimensionalidad que transforma las variables originales en nuevas combinaciones lineales (componentes principales), ordenadas según la varianza que explican en los datos, sin considerar directamente la variable objetivo.

El parámetro *nFeatures* controla cuántas características serán seleccionadas, facilitando comparaciones entre métodos bajo una misma cantidad de información.


####**Experimento "métodos de seleccion de caracteristicas"**

```python
# Prepara los métodos de selección (incluye 'passthrough' para no seleccionar)
selectors = get_feature_selection_methods(nFeatures=10)
selectors["None"] = "passthrough"

# Fija la estrategia de validación cruzada
CV = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=42)

plt.figure(figsize=(8, 6))

for name, selector in selectors.items():
    # Construye el pipeline con escalado, selección (o passtrough) y clasificador
    pipeline = ImbPipeline([
        ("scaler", MinMaxScaler()),
        ("selector", selector),
        ("clf", DecisionTreeClassifier(random_state=42))
    ])

    # Solo varía el selector en la búsqueda de hiperparámetros
    param_grid = {"selector": [selector]}
    grid = GridSearchCV(
        pipeline,
        param_grid,
        cv=CV,
        scoring="roc_auc",
        n_jobs=-1,
        verbose=0
    )
    grid.fit(X_train, y_train)

    # Obtiene probabilidades y curva ROC sobre el conjunto de prueba
    y_prob = grid.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc_score = roc_auc_score(y_test, y_prob)

    # Grafica la curva ROC de cada método
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc_score:.3f})")

# Línea de referencia de clasificador aleatorio
plt.plot([0, 1], [0, 1], "--", color="gray", alpha=0.5, label="Aleatorio")

plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curvas ROC: Influencia de la Selección de Características")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

Este fragmento de código compara visualmente distintos métodos de selección de características aplicados a un modelo de clasificación con árboles de decisión, evaluando su desempeño mediante curvas ROC.
1. Primero, se preparan tres técnicas de selección (LASSO, Anova, ExtraTrees) y una opción sin selección (None), y se define una estrategia de validación cruzada estratificada.
2. Luego, en un bucle, se construye un pipeline para cada método que incluye: escalado con MinMaxScaler, la técnica de selección correspondiente y un clasificador DecisionTreeClassifier.
3. Para cada pipeline, se ejecuta un GridSearchCV (aunque aquí no se busca hiperparámetros adicionales, solo se ajusta el modelo), se obtienen las probabilidades predichas en el conjunto de prueba, y se calcula la curva ROC y el AUC.
4. Finalmente, se grafican todas las curvas ROC en una misma figura para comparar la capacidad de discriminación del modelo según el método de selección de variables utilizado, incluyendo una línea base de clasificador aleatorio como referencia.

```python
# … (imports y definición de get_feature_selection_methods con PCA) …

selectors = get_feature_selection_methods(nFeatures=2)
full_features = df.columns.drop("Clase")
X_full = df[full_features].values
y_full = (df["Clase"] == "PNEUMONIA").astype(int).values

CV = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=42)
background_cmap = ListedColormap(['#CCE5FF', '#FFCCCC'])

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.flatten()

for ax, (name, selector) in zip(axes, selectors.items()):
    if isinstance(selector, PCA):
        # Para PCA obtenemos directamente las dos componentes principales
        X_sel = selector.fit_transform(X_full)
        feat_names = ["PC1", "PC2"]
    else:
        # Para LASSO, Anova y ET usamos get_support()
        selector.fit(X_full, y_full)
        mask = selector.get_support()
        feat_names = full_features[mask]
        X_sel = selector.transform(X_full)

    # Ahora X_sel[:, :2] es nuestro espacio 2D
    X_vis = MinMaxScaler().fit_transform(X_sel[:, :2])
    y_vis = y_full

    # Generamos la malla en ese espacio
    x_min, x_max = X_vis[:,0].min() - .1, X_vis[:,0].max() + .1
    y_min, y_max = X_vis[:,1].min() - .1, X_vis[:,1].max() + .1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )

    # Pipeline mínimo con L1-logística y mayor max_iter
    pipeline = ImbPipeline([
        ("scaler", MinMaxScaler()),
        ("clf", LogisticRegression(
            penalty='l1', solver='liblinear', C=1.0,
            max_iter=1000, random_state=42
        ))
    ])
    grid = GridSearchCV(
        pipeline,
        param_grid={"clf": [pipeline.named_steps["clf"]]},
        cv=CV,
        scoring="roc_auc",
        n_jobs=-1
    )
    grid.fit(X_vis, y_vis)

    # Predecimos la clase sobre la grilla y coloreamos el fondo
    Z = grid.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=background_cmap, alpha=0.6)

    # Superponemos los puntos reales
    ax.scatter(X_vis[y_vis==0,0], X_vis[y_vis==0,1],
               c='blue', edgecolor='k', label='Normal', s=30)
    ax.scatter(X_vis[y_vis==1,0], X_vis[y_vis==1,1],
               c='red', edgecolor='k', label='Neumonía', s=30)

    ax.set_title(name, fontsize=20)
    ax.set_xlabel(feat_names[0])
    ax.set_ylabel(feat_names[1])
    ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

Para cada caso, se entrenan modelos de regresión logística con regularización L1 mediante validación cruzada (GridSearchCV), y se visualizan los límites de decisión sobre un fondo coloreado según la clase predicha (Normal o Neumonía).
