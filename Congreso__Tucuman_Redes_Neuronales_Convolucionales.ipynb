```python
# Definimos una funci칩n que cuenta la cantidad de carpetas (clases) e im치genes dentro de un directorio ra칤z.
def contar_carpetas_e_imagenes(directorio_raiz):
    # Obtenemos la lista de carpetas que hay dentro del directorio ra칤z (cada una representa una clase)
    carpetas_clase = [d for d in os.listdir(directorio_raiz)]

    # Imprimimos el nombre del directorio que estamos analizando
    print(f"\n游늬 En el directorio: {directorio_raiz}")

    # Imprimimos la cantidad de carpetas/clases detectadas
    print(f"N칰mero de carpetas de clase: {len(carpetas_clase)}")

    total_imagenes = 0  # Inicializamos un contador para el total de im치genes

    # Recorremos cada carpeta de clase para contar cu치ntas im치genes hay en cada una
    for nombre_clase in carpetas_clase:
        # Construimos la ruta completa a la carpeta de esa clase
        dir_clase = os.path.join(directorio_raiz, nombre_clase)

        # Contamos cu치ntos archivos hay en esa carpeta (asumiendo que todos son im치genes)
        num_imagenes = len([f for f in os.listdir(dir_clase)])

        # Sumamos esa cantidad al total general
        total_imagenes += num_imagenes

    # Imprimimos el total de im치genes encontradas en el directorio ra칤z
    print(f"Total de im치genes en {directorio_raiz}: {total_imagenes}")
```

##**Imagenes**

```python
# Definimos la ruta al directorio de entrenamiento. Se construye uniendo la ruta base del dataset
# con la subcarpeta "train", que contiene las im치genes utilizadas para entrenar el modelo.
train_dir = os.path.join("/kaggle/input/chest-xray-pneumonia/chest_xray", "train")

# Definimos la ruta al directorio de validaci칩n. Esta carpeta contiene im치genes que se usan
# para evaluar el desempe침o del modelo despu칠s de cada 칠poca y ajustar hiperpar치metros.
val_dir = os.path.join("/kaggle/input/chest-xray-pneumonia/chest_xray", "val")

# Definimos la ruta al directorio de prueba (test), que contiene im치genes completamente nuevas
# que no se han usado ni en el entrenamiento ni en la validaci칩n. Se usan para evaluar la generalizaci칩n del modelo.
test_dir = os.path.join("/kaggle/input/chest-xray-pneumonia/chest_xray", "test")
```

```python
# Mostramos en pantalla las subcarpetas que existen dentro del directorio de entrenamiento.
# Por lo general, cada subcarpeta representa una clase (por ejemplo: 'NORMAL', 'PNEUMONIA').
print("Entrenamiento:", os.listdir(train_dir))

# Mostramos las subcarpetas del directorio de validaci칩n, para verificar que est칠n correctamente organizadas.
print("Validaci칩n:",   os.listdir(val_dir))

# Mostramos las subcarpetas del directorio de prueba. Esto nos ayuda a confirmar que la estructura
# del conjunto de datos est치 bien definida y lista para ser utilizada.
print("Prueba:",       os.listdir(test_dir))

```

```python
# Funci칩n para contar im치genes por clase en un directorio dado
def count_images(directory):
    counts = {}
    for clase in os.listdir(directory):
        clase_dir = os.path.join(directory, clase)
        if os.path.isdir(clase_dir):
            counts[clase] = len([f for f in os.listdir(clase_dir)
                                 if os.path.isfile(os.path.join(clase_dir, f))])
    return counts

# Contamos im치genes en cada partici칩n
train_counts = count_images(train_dir)
val_counts   = count_images(val_dir)
test_counts  = count_images(test_dir)

# Preparamos datos para el gr치fico
particiones = ['Train', 'Val', 'Test']
clases      = list(train_counts.keys())  # NORMAL, PNEUMONIA
datos = {
    'Train': [train_counts[c] for c in clases],
    'Val':   [val_counts[c]   for c in clases],
    'Test':  [test_counts[c]  for c in clases]
}

# Imprimimos un resumen
total_train = sum(train_counts.values())
total_val   = sum(val_counts.values())
total_test  = sum(test_counts.values())

print(f"Resumen de im치genes:")
print(f" - Train: {total_train} ({', '.join([f'{c}: {train_counts[c]}' for c in clases])})")
print(f" - Val:   {total_val}   ({', '.join([f'{c}: {val_counts[c]}'   for c in clases])})")
print(f" - Test:  {total_test}  ({', '.join([f'{c}: {test_counts[c]}'  for c in clases])})\n")

# Graficamos la distribuci칩n
x = np.arange(len(clases))
width = 0.25

fig, ax = plt.subplots()
ax.bar(x - width, datos['Train'], width, label='Train')
ax.bar(x,         datos['Val'],   width, label='Val')
ax.bar(x + width, datos['Test'],  width, label='Test')

ax.set_ylabel('N칰mero de im치genes')
ax.set_title('Distribuci칩n de clases por partici칩n')
ax.set_xticks(x)
ax.set_xticklabels(clases)
ax.legend()
plt.show()
```

```python
def mostrar_dos_filas_por_clase(dir_base, clases, filas_por_clase=2, cols=5, target_size=(150,150)):
    """
    Muestra filas_por_clase filas de cada clase, con cols im치genes por fila.
    """
    total_filas = filas_por_clase * len(clases)
    fig, axes = plt.subplots(
        nrows=total_filas,
        ncols=cols,
        figsize=(cols * 2.5, total_filas * 2.5)
    )

    for ci, clase in enumerate(clases):
        carpeta = os.path.join(dir_base, clase)
        archivos = [
            f for f in os.listdir(carpeta)
            if f.lower().endswith(('.png', '.jpg', '.jpeg'))
        ]
        # Elegimos filas_por_clase * cols im치genes sin repetir
        seleccion = random.sample(archivos, min(len(archivos), filas_por_clase * cols))

        for fila in range(filas_por_clase):
            imgs_fila = seleccion[fila * cols : (fila + 1) * cols]
            for col, fname in enumerate(imgs_fila):
                ax = axes[ci * filas_por_clase + fila, col]
                img = load_img(os.path.join(carpeta, fname), target_size=target_size)
                ax.imshow(img, cmap='gray')
                ax.axis('off')
                ax.set_title(f"{clase}\n{fname}", fontsize=7)

    plt.tight_layout()
    plt.show()

# Uso:
clases = ['NORMAL', 'PNEUMONIA']
mostrar_dos_filas_por_clase(train_dir, clases, filas_por_clase=2, cols=5)
```

##**Preprocesamiento**

```python
# Normalizamos p칤xeles a [0,1] y aplicamos augmentations solo en el set de entrenamiento
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

# Para validaci칩n y test, solo rescale
val_test_gen = ImageDataGenerator(rescale=1./255)

# Generadores que leen im치genes desde disco
train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=32,
    class_mode='binary'
)
val_data = val_test_gen.flow_from_directory(
    val_dir,
    target_size=(150,150),
    batch_size=32,
    class_mode='binary'
)
test_data = val_test_gen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)
```

```python

# 1. Elegimos al azar una imagen de entrenamiento de la clase NORMAL (puedes cambiar a PNEUMONIA)
clase_ejemplo = 'NORMAL'
ruta_clase = os.path.join(train_dir, clase_ejemplo)
archivo = random.choice([
    f for f in os.listdir(ruta_clase)
    if f.lower().endswith(('.png','.jpg','.jpeg'))
])
ruta_img = os.path.join(ruta_clase, archivo)

# 2. Cargamos y preparamos la imagen
img = load_img(ruta_img, target_size=(150,150))
x = img_to_array(img)            # convierte a array (150,150,3)
x = x.reshape((1,) + x.shape)    # a침ade dimensi칩n de batch, queda (1,150,150,3)

# 3. Creamos un iterador de augmentations
aug_iter = train_gen.flow(x, batch_size=1)

# 4. Mostramos 9 variaciones
plt.figure(figsize=(6,6))
for i in range(9):
    batch = next(aug_iter)       # un batch de tama침o 1
    img_aug = batch[0]           # extraemos la imagen
    plt.subplot(3,3,i+1)
    plt.imshow(img_aug)
    plt.axis('off')
    plt.title(f"{clase_ejemplo}", fontsize=8)

plt.suptitle('Ejemplos de augmentations sobre una NORMAL', fontsize=12)
plt.tight_layout()
plt.show()
```

##**Entrenamiento**

```python
model = models.Sequential([
    # Capa convolucional + pooling
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    # Aplanar y conectar densamente
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),

    # Salida binaria: normal vs neumon칤a
    layers.Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',               # Optimizador Adam: combina lo mejor de RMSProp y SGD, con ajuste adaptativo del learning rate
    loss='binary_crossentropy',     # Funci칩n de p칠rdida para clasificaci칩n binaria (ej: 0 vs 1)
    metrics=['accuracy']            # M칠trica que se reportar치 durante el entrenamiento y validaci칩n: exactitud (proporci칩n de predicciones correctas)
)
```

```python
history = model.fit(
    train_data,                    # Datos de entrenamiento (puede ser un arreglo NumPy, un generador o un objeto tf.data)
    epochs=10,                      # N칰mero de veces que el modelo ver치 *todo* el conjunto de entrenamiento
    validation_data=val_data      # Datos de validaci칩n: se usan para evaluar el desempe침o del modelo al final de cada 칠poca
)
```

##**Evaluacion del modelo**

```python
# Evaluamos el modelo ya entrenado utilizando el conjunto de prueba (test_data).
# Esto nos permite conocer c칩mo se desempe침a el modelo con datos que no ha visto antes.
test_loss, test_acc = model.evaluate(test_data)

# Imprimimos la precisi칩n obtenida en el conjunto de prueba, convertida a porcentaje
# y redondeada a dos decimales para facilitar su lectura.
print(f"\nPrecisi칩n en test: {test_acc*100:.2f}%")
```

```python
# Precisi칩n
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Precisi칩n durante el entrenamiento')
plt.xlabel('칄poca')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# P칠rdida
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('P칠rdida durante el entrenamiento')
plt.xlabel('칄poca')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
